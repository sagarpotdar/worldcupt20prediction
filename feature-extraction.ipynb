# Here we have 63888 rows 



# we will only consider city which has only hosted more than 5 matches

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

# Current Score is calcualted by cummaltive summation

   # 1. groupby columns
   # 2. use cumsum() function to do cummaltive sum of runs

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

# Calculate "ball_no"



# Steps 

   # 1. Divide over column by spliting based on dot(.) Value
   
   # Ex- 4.3over is divided into '4' over and '3' ball

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
#df
```

```python
df
```

# Calculated balls_bowled



   # Multiply 'over' column by 6 and add 'ball no' (Convert into Integer)
   
   
   
   
   # Example - 0.5 over is convert into 5balls bowled

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

# Calculating balls left 

# Simply sub 120 from balls bowled 
# and 
# Do simple Operation for negative values

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
#df
```

```python
df
```

# CALCULATE 'player dismissed' and 'wicket left'



# Steps

#   Convert player's name into '1' if present , else '0' 

   # Convert into integer format

   # Do Group by based on match id
   
   # Apply cumsum()[ie cummaltive summation] 
   
# And for 'wicket left ' do Subtract  'player dismissed' from 10

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

# Calculate 'CRR' 


                #using formula  crr=current_score*6/balls bowled  

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

#  Finding runs scored in last 5 overs

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist()) # rolling is used select range of rows to use at once.
```

```python
#last_five it has list where every match initial 30 balls is nan because we cant have 30 balls played at the starting of innings
```

```python
df['last_five'] = last_five
```

```python
#df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','']]
```

# Calculating Final Score of the match

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')




#  calculate runs scored based on match id so that match wise score can be calculated

#  reset index() is used making into dataframe

# 2 new rows will be added
```

Above operation will gives 2 new columns ie---> runs_x and runs_y  


runs_x ----> Total Runs Scored in that Match

runs_y ----> Runs Scored in every ball

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#installing xgboost if module is not found


# import sys
# !{sys.executable} -m pip install xgboost
```

# Transformation of character into integer using one hot encoding

# Rest Integer are left passed

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

# Defining Pipeline of Sklearn



# 1st Step ----> Transformation
# 2nd Step-----> Standard Scaling
# 3rd Step -----> Using XG Boost Regressor With Changing Hyper-Parameter


```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)                           # Training Model 
y_pred = pipe.predict(X_test)                       # Prediction of Model

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
